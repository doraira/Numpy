# np.copy(a, order='K', subok=True)


```python
"""
➤ np.copy()
تستخدم لإنشاء نسخة مستقلة تمامًا 
(deep copy) 
من المصفوفة الأصلية.
يعني:

التعديلات على النسخة الجديدة لا تؤثر على الأصل.

ودي حاجة مهمة لو مش عايزة تخسري البيانات الأصلية.



"""
```

```python
a = np.array([1, 2, 3])
b = np.copy(a)

b[0] = 99

print("Original:", a)  
print("Copy:    ", b) 
```

```python
# تعديل b لم يؤثر على a.

```

```python
#❗ الفرق بين copy() و = (assignment)
```

```python
a = np.array([1, 2, 3])
b = a        # هذا مجرد مرجع (reference)

b[0] = 99

print("a:", a)  # ➤ [99 2 3] ← تأثر!
```

```python
"""
b = a ← مجرد مرجع لنفس المكان في الذاكرة.

b = np.copy(a) ← نسخة مستقلة تمامًا.
"""
```

```python
"""
🔄 الفرق بين 
copy() 
و 
deepcopy() 
في بايثون العادي:
np.copy() = deep copy لمصفوفة numpy فقط.

copy.deepcopy() = نسخ عميق لأي كائن في بايثون (حتى المتداخلين).
"""
```

```python
"""
🔄 الفرق بين 
np.copy() 
و
copy.deepcopy() 
في التعامل مع ده:
1. np.copy():
تنسخ مصفوفة 
numpy 
بس
حتى لو جواها أشياء متداخلة، مش دايمًا تقدر تتعامل مع كل الأعماق.

2. copy.deepcopy():
من مكتبة 
copy 
في بايثون.

بتعمل نسخ عميق لكل حاجة جوا الكائن مهما كان متداخل.
"""
```

```python
import copy

original = [[1, 2], [3, 4]]
shallow = original.copy()

shallow[0][0] = 999

print(original)  # ➤ [[999, 2], [3, 4]] ← اتغيّر الأصل!
```

```python
deep = copy.deepcopy(original)

deep[0][0] = 111

print(original)  # ➤ [[1, 2], [3, 4]] ← سليم، مفيش تغيّر
print(deep)      # ➤ [[111, 2], [3, 4]]
```

```python
"""
مشروع تحليل بيانات طلاب، وفيه مصفوفة 
Numpy 
اسمها 
grades_array
استخدام 
np.copy()
او
copy.deepcopy()
"""


```

```python
"""
بيانات بسيطة من نوع 
NumPy Array 
(غير متداخلة)
نستخدم 
np.copy()
"""
```

```python
grades_array = np.array([80, 90, 70, 85])

# نعمل نسخة منها للتعديل:
copy_grades = np.copy(grades_array)

copy_grades[0] = 100

print("Original:", grades_array)  
print("Copy:", copy_grades)    
```

```python
"""
بيانات متداخلة 
(قائمة جواها مصفوفات – Nested)
"""
```

```python
nested_data = [[np.array([80, 90])], [np.array([70, 85])]]

# shallow copy (مش هتنفع هنا)
shallow = nested_data.copy()

shallow[0][0][0] = 999  # تعديل العنصر الأول

print(nested_data)  # ❌ الأصل اتغير!
```

```python
deep = copy.deepcopy(nested_data)
deep[0][0][0] = 555

print(nested_data)  # ✅ سليم، مفيش تغيير
print(deep)         # ✅ النسخة اتعدلت لوحدها
```

# np.random.permutation(x)


```python
"""
np.random.permutation() 
ترجع نسخة مرتبة عشوائيًا 
(shuffled) من:

أرقام من 0 إلى 
n-1 
لو أعطيتها عددًا صحيحًا 
n.
أو نسخة 
shuffled من مصفوفة/قائمة/
array موجودة لو أعطيتها مصفوفة
"""
```

```python
"""
✅ np.random.permutation()
بترجع نسخة جديدة من البيانات 
shuffled.
الأصل ما بيتغيرش.
✅ np.random.shuffle()
ما بترجعش حاجة 
(ترجع None).
لكن بتغير البيانات الأصلية نفسها 
(in-place).
"""
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([1, 2, 3, 4])

# permutation → نسخة جديدة
shuffled1 = np.random.permutation(arr1)
print("Original arr1:", arr1)        

print("Shuffled copy:", shuffled1)   


# shuffle → تعديل في الأصل
np.random.shuffle(arr2)
print("Shuffled in-place arr2:", arr2)  

```

```python
# 🔹الهدف: إنشاء ترتيب عشوائي بدون تكرار من مجموعة أرقام.

np.random.permutation(5)
```

```python
arr = np.array([10, 20, 30, 40])
shuffled = np.random.permutation(arr)
print(shuffled)
```

```python
# توزيع الأسئلة أو المهمات عشوائيًا

questions = ["Q1", "Q2", "Q3", "Q4"]
np.random.permutation(questions)
```

```python
# تحضير بيانات التدريب في ML:
X = np.array([[1], [2], [3], [4]])
X_shuffled = np.random.permutation(X)
X_shuffled
```

# np.argwhere(condition)


```python
"""
تعيد إحداثيات 
(indices) 
العناصر التي تحقق شرط معين داخل مصفوفة 
numpy
"""
```

```python
import numpy as np

a = np.array([[10, 20], [30, 15]])

# نجيب أماكن العناصر الأكبر من 15
indices = np.argwhere(a > 15)

print(indices)
```

```python
"""
[0, 1] ← العنصر 20

[1, 0] ← العنصر 30
"""
```

```python
# 🤔 طيب ليه بنكتب [0][0] في بعض الأحيان؟
```

```python
idx = np.argwhere(a == 20)[0][0]
```

```python
idx # row index only
```

```python
"""
📌 التفسير:

np.argwhere(a == 20) → [[0, 1]] (2D array)

[0] → أول نتيجة: [0, 1]

[0][0] → العنصر الأول في [0, 1] → 0 (الـ row index فقط)
"""
```

```python
"""
🚨 ملاحظات:
النتيجة دائمًا تكون 2D حتى لو كان a 1D.

لو مفيش ولا عنصر بيحقق الشرط → النتيجة هتكون مصفوفة فاضية: array([], shape=(0, 2), dtype=int64)

"""
```

```python
"""
أنتِ بتشتغلي على تحليل بيانات طبية (مثلاً بيانات ضغط الدم لمجموعة من المرضى)، وعايزة تجهزي البيانات قبل ما تدربي نموذج تنبؤي
🎯 الهدف:
تحددي أماكن القيم المفقودة (NaN) أو

القيم المتطرفة (outliers) علشان:

تحذفيها

أو تعوضيها

أو تتخذي قرار بتنظيفها
"""
```

```python
data = np.array([[120, 125], [np.nan, 140], [135, np.nan]])

# نستخدم np.isnan() عشان نحدد أماكن الـ NaN
nan_indices = np.argwhere(np.isnan(data))

print("Missing values found at:", nan_indices)
```

```python
# تعريف: أي قيمة أكبر من 180 نعتبرها outlier
outliers = np.argwhere(data > 180)

print("Outliers at:", outliers)
```
